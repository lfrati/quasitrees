```python
from trees import prufer2edges, edges2tree, visit, all_prufers
import networkx as nx
import matplotlib.pyplot as plt
from math import ceil
```

```python
# Each prufer sequence of length N uniquely identifies a labelled tree of length N+2
```

```python
prufer = [4, 4, 4, 5]
edges = prufer2edges(prufer)

G = nx.Graph(edges)
nx.draw(G, with_labels=True, font_weight="bold", node_color="white", edgecolors="black")
```

```python
# the set of edges defined by the prufer sequence can be converted to a tree and then
# visited to generate a permutation
```

```python
tree = edges2tree(edges)
s = visit(tree, 1)
print(s)
```

```python
# each tree correponds to multiple permutations depending on root placement
tree = edges2tree(edges)
for root in sorted(tree.keys()):
    s = visit(tree, root)
    print(f"root: {root} perm: {s}")

# and visit order
# TODO
```

```python
# for N=5 nodes there are N**(N-2) = 5**3 = 125 prufer sequences
N = 5
prufers = all_prufers(N)
side = ceil(len(prufers)**0.5)
fig, axs = plt.subplots(ncols=side, nrows=side, figsize=(14, 14))

for ax in axs.flatten():
    ax.axis("off") # hide extra cells
    
for p, ax in zip(prufers, axs.flatten()):
    edges = prufer2edges(p)
    prufer_code = "".join(map(str, p))
    G = nx.Graph(edges)
    nx.draw(
        G,
        ax=ax,
        with_labels=True,
        font_weight="bold",
        node_size=80,
        node_color="white",
        font_size=6,
    )
    ax.set_title(prufer_code, fontsize=10)
plt.tight_layout()
```
